### Dataset Overview
Your dataset contains **~160,000 healthcare encounters** from **25,000 unique patients** with **35 variables**:

**Key Structure:** Each patient has **multiple visits** (average ~6 visits per patient). The same `patient_id` appears multiple times with different `service_date` values - this is NOT a duplicate error!

**Identifiers & Dates:**
- `patient_id` - Patient identifier (same patient has multiple rows)
- `date_of_birth` - Patient date of birth (constant per patient)
- `service_date` - Date of healthcare encounter (varies per visit)

**Demographics (constant per patient):**
- `sex` - Male/Female (with typos like 'Mael', 'm', 'MALE', etc.)
- `ethnicity` - 8 categories
- `state` - US state (30 states)

**Clinical Measures (vary per visit):**
- `height_cm`, `weight_kg` - Height/weight (calculate BMI yourself!)
- `systolic_bp`, `diastolic_bp` - Blood pressure readings
- `heart_rate` - Resting heart rate (bpm)
- `total_cholesterol`, `hdl_cholesterol`, `ldl_cholesterol` - Lipid panel (~40% missing - not every visit has labs)
- `fasting_glucose` - Blood sugar (mg/dL)

**Behavioral Factors:**
- `smoking_status` - Never/Former/Current
- `diabetes_status` - No/Yes/Pre-diabetic (with MANY typos!)
- `physical_activity_min`, `alcohol_drinks_week`

**Medical Codes (KEY FOR ANALYSIS):**
- `procedure_code` - CPT codes (99213=office visit, 92920=PCI, 33533=CABG, etc.)
- `diagnosis_code` - ICD-10 codes, comma-separated (format: char5, e.g., 'I6523')
- `medication_list` - Comma-separated list of current medications
- `family_history_cvd` - Family history of CVD (0/1)

**Social Determinants:**
- `education_level`, `income_bracket`, `insurance_type`
- `sleep_hours`, `stress_level`

**Noise Variables (irrelevant - practice feature selection!):**
- `favorite_color`, `pet_ownership`, `birth_month`
- `handedness`, `num_siblings`, `eye_color`

**Outcome (YOU CREATE THIS!):**
- `cvd_risk_score` - Continuous risk score (0-100) - for reference
- **NO `cvd_event` column!** You must identify CVD events from `diagnosis_code` using ICD-10 codes:
  - Cerebrovascular: I60-I69 (stroke, TIA, cerebral atherosclerosis)
  - Cardiovascular: I25.x (coronary artery disease), I50.x (heart failure), I73.x (PVD)
